# Goal

```python
def goals(mma):
    df=pd.DataFrame()
#     for i in list (mma.Team.unique()):
#         byteam=mma[mma['Team']==i]
    shot=mma[(mma["Event"]=="Shot")|(mma["Event"]=="Goal")]
    goal=mma[(mma["Event"]=="Goal")]
    shot=shot.groupby(['Player']).count().reset_index()
    goal=goal.groupby(['Player']).count().reset_index()
    shot['total_shots']=shot['Event']
    goal['total_goal']=goal['Event']
    shot=shot[['Player','total_shots']]
    goal=goal[['Player','total_goal']]
    df1=pd.merge(mma,shot,on = "Player")
    df2=pd.merge(df1,goal, on="Player")
    df2['goal_proportion']=np.where(df2['total_goal']==0,0,df2['total_goal']/df2['total_shots'])
    df2=df2.sort_values('goal_proportion',ascending = True)
    df2['Rank_goal']=df2['goal_proportion'].rank(method='dense',ascending = True)
#     df=df.append(df2)

    df2.drop(columns=['total_goal','total_shots'],inplace=True)
    df2.reset_index(inplace=True)
    df1=df2[['Player','goal_proportion','Team','Rank_goal']]
    df1.drop_duplicates(inplace=True)
    print(df1.head())
    print(len(df1.Player.unique()))
    return df1
```

```python
df1=goals(mma)
```

# Takeaway Specialist

```python
def takeaway(mma):
    df=pd.DataFrame()
#     for i in list (mma.Team.unique()):
#         byteam=mma[mma['Team']==i]
    takeaway=mma[(mma["Event"]=="Takeaway")]
    takeaway=takeaway.groupby(['Player']).count().reset_index()
    takeaway['total_takeaways']=takeaway['Event']
    takeaway=takeaway[['Player','total_takeaways']]
    df2=pd.merge(mma,takeaway,on = "Player")
    df2=df2.sort_values('total_takeaways',ascending = True)
    df2['Rank_takeaway']=df2['total_takeaways'].rank(method='dense',ascending = True)
#     df=df.append(df2)

    df2.reset_index(inplace=True)
    df4=df2[['Player','total_takeaways','Team','Rank_takeaway']]
    df4.drop_duplicates(inplace=True)
    print(df4)
    return df4
```

# Passer

```python
def passer(data):
    df=pd.DataFrame()
#     for i in list (data.Team.unique()):
#         byteam=data[data['Team']==i]
    passes=data[(data["Event"]=="Play")|(data["Event"]=="Incomplete Play")]
    play=data[(data["Event"]=="Play")]
    passes=passes.groupby(['Player']).count().reset_index()
    play=play.groupby(['Player']).count().reset_index()
    passes['total_passes']=passes['Event']
    play['successful_passes']=play['Event']
    passes=passes[['Player','total_passes']]
    play=play[['Player','successful_passes']]
    df1=pd.merge(data,passes,on = "Player")
    df2=pd.merge(df1,play, on="Player")
    df2['pass_proportion']=np.where(df2['successful_passes']==0,0,df2['successful_passes']/df2['total_passes'])
    df2=df2.sort_values('pass_proportion',ascending = True)
    df2['Rank_pass']=df2['pass_proportion'].rank(method='dense',ascending = True)
#     df=df.append(df2)
    
    df2.reset_index(inplace=True)
    df2.drop(columns=['successful_passes','total_passes'],inplace=True)
    df2=df2[['Player','pass_proportion','Team','Rank_pass']]
    df2.drop_duplicates(inplace=True)
    print(df2.tail())
    return df2
```

```python
df3=passer(data)
```

# Faceoff Wins

```python
def faceoff_wins(data):
    df=pd.DataFrame()
#     for i in list (data.Team.unique()):
#         byteam=data[data['Team']==i]
    faceoff=data[(data["Event"]=="Faceoff Win")]
    faceoff=faceoff.groupby(['Player']).count().reset_index()
    faceoff['total_wins']=faceoff['Event']
    faceoff=faceoff[['Player','total_wins']]
    df2=pd.merge(data,faceoff,on = "Player")
    df2=df2.sort_values('total_wins',ascending = True)
    df2['Rank_faceoffWin']=df2['total_wins'].rank(method='dense',ascending = True)
#     df=df.append(df2)

    df3=df2[['Player','total_wins','Team','Rank_faceoffWin']]
    df3.drop_duplicates(inplace=True)
#     df3.reset_index(inplace=True)
    print(df3.tail())
    return df3
```

```python
df4=faceoff_wins(data)
```

# Faceoff Losses

```python
def faceoff_loss(data):
    df=pd.DataFrame()
    faceoff=data[(data["Event"]=="Faceoff Win")]
    faceoff=faceoff.groupby(['Player 2']).count().reset_index()
    faceoff['total_losses']=faceoff['Event']
    faceoff=faceoff[['Player 2','total_losses']]
    df2=pd.merge(data,faceoff,left_on = "Player", right_on='Player 2')
#     for i in list (df2.Team.unique()):
#         byteam=df2[df2['Team']==i]
    df2=df2.sort_values('total_losses',ascending = True)
    df2['Rank_faceoffLoss']=df2['total_losses'].rank(method='dense',ascending = False)
#     df=df.append(byteam)

    df3=df2[['Player','total_losses','Team','Rank_faceoffLoss']]
    df3.drop_duplicates(inplace=True)
#     df3.reset_index(inplace=True)
    print(df3.head())
    return df3
```

```python
df5=faceoff_loss(data)
```

```python
data1=data.copy()
data1=data1[['Player','Team']]
df9=pd.merge(df1,data1,how='right')
df6=pd.merge(df9,df2,on=['Player','Team'],how='left')
df7=pd.merge(df6,df3,on=['Player','Team'],how='left')
df8=pd.merge(df7,df4,on=['Player','Team'],how='left')
final_df=pd.merge(df8,df5,on=['Player','Team'],how='left')
# final_df=final_df.drop(columns=['index','index_x','index_y'])
final_df.drop_duplicates(inplace=True)
# final_df.to_excel('final.xlsx')
final_df.head()
```
